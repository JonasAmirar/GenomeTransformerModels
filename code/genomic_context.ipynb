```r
snv_path =   '/omics/groups/OE0606/internal/jonas/genomic_data/snv_variant_calls'
ref_genome = '/omics/groups/OE0606/internal/shared/icgc_pcawg/reference_genome/GRCh37d5/genome.fa'

context_size <- 255
```

```r
# Read the list of files from shared_files.txt
shared_files <- readLines('/omics/groups/OE0606/internal/jonas/genomic_data/shared_files.txt')
snv_files <- file.path(snv_path, paste0(shared_files, ".csv"))
```

```r
for (file in snv_files) {
    print(file)
    print('.')
    dat <- read.csv(file, stringsAsFactors = FALSE)
    
    # Get sequence context
    context=scanFa(file=ref_genome, resize(makeGRangesFromDataFrame(dat, seqnames.field="X.CHROM", start.field="POS", end.field="POS"), context_size, fix="center"))
    
    # Add the x column from context_df to dat 
    dat[[paste0("seq_context_", context_size)]] <- as.data.frame(context)$x
    write.csv(dat, file=file, row.names=FALSE)
}
```

```r
for (file in snv_files) {
    #print(file)
    #print('.')
    
    # Check if the file already has the seq_context_255 column
    dat <- read.csv(file, stringsAsFactors = FALSE)
    context_col_name <- paste0("seq_context_", context_size)
    
    if (!(context_col_name %in% colnames(dat))) {
        tryCatch({
            # Get sequence context
            context <- scanFa(file=ref_genome, resize(makeGRangesFromDataFrame(dat, seqnames.field="X.CHROM", start.field="POS", end.field="POS"), context_size, fix="center"))
            
            # Add the x column from context_df to dat 
            dat[[context_col_name]] <- as.data.frame(context)$x
            write.csv(dat, file=file, row.names=FALSE)
        }, error = function(err) {
            # Log the error
            print(paste("Error processing file:", conditionMessage(err)))
        })
    } else {
        print(paste("Column", context_col_name, "already exists"))
    }
}
```
