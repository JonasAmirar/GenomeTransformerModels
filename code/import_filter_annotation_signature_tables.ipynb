```python
def snv_gz_to_csv(snv_file_name, verbose=False):
    
    snv_file_path = f"{snv_path}/{snv_file_name}"

    # Find the index of the last header line (starting with a single '#')
    header_line = -1
    with gzip.open(snv_file_path, 'rt') as file:  # Open as text with gzip handling
        for i, line in enumerate(file):
            if line.startswith('#'):
                header_line = i
    
    # Use the identified header line to read the VCF data into a DataFrame
    snv_df = pd.read_csv(snv_file_path, compression='gzip', header=header_line, sep='\t', dtype={0: str})
    if verbose:
        print('Original data', '\n',  snv_df.head(3), '\n\n\n\n')
    snv_df = snv_df[['#CHROM', 'POS', 'REF', 'ALT', 'INFO']]
    if verbose:
        print('Relevant columns', '\n',snv_df.head(3),'\n\n\n\n')


    # Check if 'REF' or 'ALT' have more than one character
    condition = (snv_df['REF'].str.len() > 1) | (snv_df['ALT'].str.len() > 1)
    # Count the number of such rows
    count_multichars = condition.sum()
    if count_multichars!=0:
        print(f"Number of rows with multiple characters in 'REF' or 'ALT': {count_multichars}")
    # Drop the rows where 'REF' or 'ALT' have more than one character
    snv_df = snv_df[~condition]

    # Define the regular expression for extracting the needed information
    pattern = r'VAF=(?P<VAF>\d+\.\d+);.*t_alt_count=(?P<t_alt_count>\d+);t_ref_count=(?P<t_ref_count>\d+);Variant_Classification=(?P<Variant_Classification>[^;]+)'

    # Extract the data using the regex pattern
    info_extracted = snv_df['INFO'].str.extract(pattern)

    # Convert extracted data to numeric where necessary (for VAF, t_alt_count, and t_ref_count)
    info_extracted['VAF'] = pd.to_numeric(info_extracted['VAF'])
    info_extracted['t_alt_count'] = pd.to_numeric(info_extracted['t_alt_count'])
    info_extracted['t_ref_count'] = pd.to_numeric(info_extracted['t_ref_count'])

    # Add the extracted columns to the original DataFrame
    snv_small = pd.concat([snv_df, info_extracted], axis=1)
    
    # Optionally, drop the original 'INFO' column if no longer needed
    snv_small.drop('INFO', axis=1, inplace=True)
    if verbose:
        print('Extracted info', '\n',snv_small.head(3))
    return snv_small
```

```python
import os

gz_path  = '/omics/groups/OE0606/internal/shared/icgc_pcawg/variant_calls/final_consensus_12oct_passonly/snv_mnv'
csv_path = '/omics/groups/OE0606/internal/jonas/genomic_data/snv_variant_calls'

for filename in os.listdir(gz_path):
    if filename.endswith(".vcf.gz"):
        csv_filename = filename.replace('.consensus.20160830.somatic.snv_mnv.vcf.gz', '.csv')
        csv_filepath = os.path.join(csv_path, csv_filename)
        
        # Convert VCF to CSV using your existing function
        snv_df = snv_gz_to_csv(filename)
        
        # Save the DataFrame to a CSV file
        snv_df.to_csv(csv_filepath, index=False)
```
